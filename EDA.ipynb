```python
%pip install opencv-python
%pip install pandas
%pip install matplotlib
import cv2 as cv
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
from IPython.display import display, Image
directory = r"D:\New Downloads\facial-emotion-dataset\train"
categories = ["angry", "disgust", "fear", "happy", "neutral", "sad", "surprise"]
df = pd.DataFrame();
for c in categories:
    path = os.path.join(directory, c)
    print(c)
    for img in os.listdir(path):
        image = cv.imread(os.path.join(path, img), cv.IMREAD_GRAYSCALE)
        new_arr = cv.resize(image, (28, 28));
        df2 = pd.DataFrame([{"arr": new_arr, "label": c}])
        df = df.append(df2);
```

```python
df
```

```python
from sklearn import preprocessing
from keras.utils import to_categorical

df['label'] = to_categorical(df['label'])
```

```python
from keras.models import Sequential
from keras.layers import Dense, Flatten, Conv2D
from sklearn.model_selection import train_test_split


X = np.array(df["arr"].to_list())
y = np.array(df["label"].to_list())

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.4,random_state=30)
```

```python
print(X_train.shape)
print(y.shape)
print(type(X_train))
print(type(y_train))
```

```python
model = Sequential()

model.add(Conv2D(64, kernel_size = 3, activation = 'relu', input_shape = X_train.shape[1:]))
model.add(Conv2D(32, kernel_size = 3, activation = 'relu'))
model.add(Flatten())
model.add(Dense(7, activation='softmax'))

model.compile(optimizer='adam', loss ='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, to_categorical(y_train), epochs=5, validation_data=(X_test, to_categorical(y_test)))
```
